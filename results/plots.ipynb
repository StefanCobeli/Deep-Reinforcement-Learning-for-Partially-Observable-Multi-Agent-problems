```python
import numpy as np
from scipy import stats

%matplotlib inline
import matplotlib as mpl
from matplotlib import rc
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
sns_colors = sns.color_palette(None, 8)
sns_colors = ['#1075a1','#24b476','#e94d5b','#7D48DA','#faa44f','#1fb0e7']

# rc('text', usetex=True)
# plt.rc('font', family='serif')
# import matplotlib as mpl
rc('font', family='serif')
# rc('font', serif='Times')
# rc('font', serif='CMU Serif')
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

# Hats riddle plots

```python
data     = np.loadtxt("statistics/hats_riddle/average_saved_agents_3")

x_axis   = np.arange(1, len(data) + 1)

data    /= 2.5
acc_mean_3 = pd.rolling_mean(data, 200, min_periods=1)#data
acc_std_3  = pd.rolling_std(data, 10000, min_periods=1)


data     = np.loadtxt("statistics/hats_riddle/average_saved_agents_4")
data    /= 3.5
acc_mean_4 = pd.rolling_mean(data, 200, min_periods=1)#data
acc_std_4  = pd.rolling_std(data, 10000, min_periods=1)

data     = np.loadtxt("statistics/hats_riddle/average_saved_agents_5")
data    /= 4.5
acc_mean_5 = pd.rolling_mean(data, 200, min_periods=1)#data
acc_std_5  = pd.rolling_std(data, 10000, min_periods=1)


data     = np.loadtxt("statistics/hats_riddle/average_saved_agents_8")
data    /= 7.5
acc_mean_8 = pd.rolling_mean(data, 200, min_periods=1)#data
acc_std_8  = pd.rolling_std(data, 10000, min_periods=1)


data     = np.loadtxt("statistics/hats_riddle/average_saved_agents_10")
data    /= 9.5 
acc_mean_10 = pd.rolling_mean(data, 200, min_periods=1)#data
acc_std_10  = pd.rolling_std(data, 10000, min_periods=1)
```

```python
fig, ax = plt.subplots(figsize=(8.8, 5.5))
# plt.title(r'Reward vs Information Entropy', fontsize=18)
plt.plot(x_axis, acc_mean_3, color=sns_colors[0], linestyle='-', linewidth=3, label="3 prisoners accuracy")
plt.fill_between(x_axis, acc_mean_3 - acc_std_3, acc_mean_3 + acc_std_3, facecolor=sns_colors[0], edgecolor="none", alpha=0.1)

plt.plot(x_axis, acc_mean_4, color=sns_colors[1], linestyle='-', linewidth=3, label="4 prisoners accuracy")
plt.fill_between(x_axis, acc_mean_4 - acc_std_4, acc_mean_4 + acc_std_4, facecolor=sns_colors[1], edgecolor="none", alpha=0.1)

plt.plot(x_axis, acc_mean_5, color=sns_colors[2], linestyle='-', linewidth=3, label="5 prisoners accuracy")
plt.fill_between(x_axis, acc_mean_5 - acc_std_5, acc_mean_5 + acc_std_5, facecolor=sns_colors[2], edgecolor="none", alpha=0.1)

plt.plot(x_axis, acc_mean_8, color=sns_colors[3], linestyle='-', linewidth=3, label="8 prisoners accuracy")
plt.fill_between(x_axis, acc_mean_8 - acc_std_8, acc_mean_8 + acc_std_8, facecolor=sns_colors[3], edgecolor="none", alpha=0.1)

plt.plot(x_axis, acc_mean_10, color=sns_colors[4], linestyle='-', linewidth=3, label="10 prisoners accuracy")
plt.fill_between(x_axis, acc_mean_10 - acc_std_10, acc_mean_10 + acc_std_10, facecolor=sns_colors[4], edgecolor="none", alpha=0.1)



plt.ylabel(r'')
plt.xlabel(r'# Episodes', fontsize=18)
ax.get_xaxis().set_major_formatter(mpl.ticker.FuncFormatter(lambda x, p: str(int(x/1000.))+'k' if x > 0 else '')) #format(int(x), ',')))
plt.yticks(np.arange(0.3, 1.1, .1))
plt.xlim(x_axis[0], x_axis[-1])
plt.ylim(0.45, 1.05)
plt.legend(borderaxespad=0.2, fontsize=12, loc=4, bbox_to_anchor=(1, 1.01), ncol=3)#(1.2, 0)
# plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.tick_params(axis='both', which='major', labelsize=16)
plt.tick_params(axis='both', which='minor', labelsize=8)
plt.tight_layout()
plt.savefig('plots/DDRQNHatsRiddleResults.pdf', bbox_inches='tight', dpi=20)
```
